### **1. Import Library**

```python
import matplotlib.pyplot as plt
import numpy as np
```

### **2. Dataset**

```python
# Dataset
stores = ['Store A','Store B','Store C']
male = [150,200,180]
female = [120,140,170]

# Data quarter
q1_male = [150,180,160]
q1_female = [140,200,180]
q2_male = [170,190,175]
q2_female = [130,210,160]
q3_male = [160,180,150]
q3_female = [135,195,165]
q4_male = [180,200,190]
q4_female = [150,170,160]
```

### **3. Basic Stacked Vertical Bar Chart**

```python
x = np.arange(len(stores))
plt.bar(x, male, label='Male', color='blue')
plt.bar(x, female, bottom=male, label='Female', color='pink')

plt.title('Population by Gender & Store')
plt.xlabel('Stores')
plt.ylabel('Population')
plt.xticks(x, stores)
plt.legend()
plt.show()
```

### **4. Kustomisasi Stacked Vertical Bar Chart**

```python
x = np.arange(len(stores))
plt.bar(x, male, label='Male', color='blue')
plt.bar(x, female, bottom=male, label='Female', color='pink')

plt.title('Population by Gender & Store')
plt.xlabel('Stores')
plt.ylabel('Population')
plt.xticks(x, stores)
plt.legend()

for i in range(len(x)):
    plt.text(x[i], male[i]/2, str(male[i]), ha='center', color='white')
    plt.text(x[i], male[i] + female[i]/2, str(female[i]), ha='center', color='black')

plt.show()
```

### **5. Multiple Stacked Vertical Bar Chart**

```python
width = 0.2
x = np.arange(len(stores))

plt.bar(x-width*1.5, q1_male, label='Q1 Male', color='grey', width=width)
plt.bar(x-width*1.5, q1_female, bottom=q1_male, label='Q1 Female', color='violet', width=width)

plt.bar(x-width/2, q2_male, label='Q2 Male', color='lightblue', width=width)
plt.bar(x-width/2, q2_female, bottom=q2_male, label='Q2 Female', color='pink', width=width)

plt.bar(x+width/2, q3_male, label='Q3 Male', color='lightgreen', width=width)
plt.bar(x+width/2, q3_female, bottom=q3_male, label='Q3 Female', color='coral', width=width)

plt.bar(x+width*1.5, q4_male, label='Q4 Male', color='blue', width=width)
plt.bar(x+width*1.5, q4_female, bottom=q4_male, label='Q4 Female', color='purple', width=width)

plt.title('Population by Gender & Store (Multiple Quarters)')
plt.xlabel('Stores')
plt.ylabel('Population')
plt.xticks(x, stores)
plt.legend(bbox_to_anchor=(1.05, 1),loc='upper left')

plt.show()
```
