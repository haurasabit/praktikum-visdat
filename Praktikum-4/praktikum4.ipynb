**a. Basic Bar Chart dengan Matplotlib**

```python
import matplotlib.pyplot as plt

#data
jurusan = [
    'Teknik Informatika',
    'Sistem Informasi',
    'Ilmu Komputer',
    'Data Science',]

jumlah_mahasiswa = [120,150,100,80]

#membuat basic bar chart
plt.bar(jurusan, jumlah_mahasiswa, color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa per Jurusan')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**b. Kustomisasi Basic Bar Chart**

```python
import matplotlib.pyplot as plt

#data
jurusan = [
    'Teknik Informatika',
    'Sistem Informasi',
    'Ilmu Komputer',
    'Data Science',]

jumlah_mahasiswa = [90,50,100,80]

#membuat basic bar chart
plt.bar(jurusan, jumlah_mahasiswa, color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa per Jurusan')
plt.xticks(rotation=45)

#menambahkan label pada batang
for i, v in enumerate(jumlah_mahasiswa):
    plt.text(i, v+1, str(v), ha='center',)

plt.show()
```

**c. Multiple Bar Chart**

```python
import matplotlib.pyplot as plt

#data
jurusan = [
    'Teknik Informatika',
    'Sistem Informasi',
    'Ilmu Komputer',
    'Data Science',]

jumlah_mahasiswa = [120,150,100,80]
jumlah_mahasiswa_2024 = [100,120,90,70]

x = range(len(jurusan))
width = 0.4

plt.bar(x, jumlah_mahasiswa, width, label='2023', color='purple')
plt.bar([p + width for p in x], jumlah_mahasiswa_2024, width=width, label='2024', color='magenta')

plt.title('Jumlah Mahasiswa per Jurusan (2023 vs 2024)')
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)
plt.xticks([p + width/2 for p in x], jurusan)
plt.legend()

plt.show()
```

**d. Basic Bar Chart untuk Melihat Pola & Tren**

```python
import matplotlib.pyplot as plt

#data
jurusan = ['Teknik Informatika','Ilmu Komputer','Sistem Informasi','Data Science',]
tahun = ['2019','2020','2021','2022','2023']
data = {
    'Ilmu Komputer': [100,110,120,130,140],
    'Teknik Informatika': [80,90,100,110,120],
    'Sistem Informasi': [70,80,90,100,110],
    'Data Science': [60,70,80,90,100]}

width = 0.2
x = range(len(tahun))

#visualisasi masing-masing jurusan
plt.bar(x, data['Ilmu Komputer'], width=width, label='Ilmu Komputer', color='lightblue')
plt.bar([p + width for p in x],
        data['Teknik Informatika'], width=width, label='Teknik Informatika', color='lightgreen')
plt.bar([p + width*2 for p in x],
        data['Sistem Informasi'], width=width, label='Sistem Informasi', color='lightpink')
plt.bar([p + width*3 for p in x],
        data['Data Science'], width=width, label='Data Science', color='lightcoral')

#detail grafik
plt.title('Jumlah Mahasiswa per Jurusan (2019-2023)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)
plt.xticks([p + width for p in x], tahun)
plt.legend()

plt.show()
```
